```python
import georasters
```

```python
#help(rasterstats.io.rasterio.open) 
raster = georasters.from_file(r"c:\scidb\glc2000_clipped.tif")
```

```python
plt.figure(figsize=(12, 8))
#raster.plot()
```

```python

minx, miny, maxx, maxy = raster.bounds
from shapely.geometry import box, mapping
myPoly = box(minx, miny, maxx, maxx)
myPoly.to_wkt()
#minx, miny, maxx, maxy

import fiona

# Here's an example Shapely geometry
#poly = Polygon([(0, 0), (0, 1), (1, 1), (0, 0)])

# Define a polygon feature geometry with one attribute
schema = {
    'geometry': 'Polygon',
    'properties': {'id': 'int'},
}

# Write a new Shapefile
with fiona.open(r'c:\work\my_shp2.shp', 'w', 'ESRI Shapefile', schema) as c:
    ## If there are multiple geometries, put the "for" loop here
    c.write({
        'geometry': mapping(myPoly),
        'properties': {'id': 123},
    })
    
#from shapely.geometry import Polygon
#polygon = Polygon([(0, 0), (1, 1), (1, 0)])
```
